```python
import numpy as np 
import sympy as sp 
import matplotlib.pyplot as plt

sp.init_printing()
```

```python
# Cell 3: Basic Derivatives (Markdown)
"""
## 1) Basic Derivatives with SymPy
Let's define a variable and a function, then compute its derivative.
"""
```

```python
# Cell 4: Basic Derivative Example
x = sp.Symbol('x')
f = x**2 + 3*x + 5
f_prime = sp.diff(f, x)

f, f_prime
```

```python
# Cell 5: Rules of Differentiation (Markdown)
"""
## 2) Rules of Differentiation
Examples of power, product, quotient, and chain rule.
"""
```

```python
# Cell 6: Rules Examples
# Power rule: d/dx (x^4)
power_rule = sp.diff(x**4, x)

# Product rule: d/dx (x^2 * sin(x))
prod_rule = sp.diff(x**2 * sp.sin(x), x)

# Quotient rule: d/dx (x^2 / (x+1))
quot_rule = sp.diff(x**2 / (x+1), x)

# Chain rule: d/dx (sin(x^2))
chain_rule = sp.diff(sp.sin(x**2), x)

power_rule, prod_rule, quot_rule, chain_rule
```

```python
# Cell 7: Visualizing Derivatives (Markdown)
"""
## 3) Visualizing Derivatives as Slopes
We plot a function and its tangent line at a given point.
"""
```

```python
# Cell 8: Visualization
f = x**2
f_prime = sp.diff(f, x)

# Convert to numerical functions
f_lambd = sp.lambdify(x, f, 'numpy')
f_prime_lambd = sp.lambdify(x, f_prime, 'numpy')

# Choose a point
x0 = 2
y0 = f_lambd(x0)
slope = f_prime_lambd(x0)

# Tangent line: y = slope*(x-x0) + y0
x_vals = np.linspace(-3, 3, 200)
y_vals = f_lambd(x_vals)
tangent = slope * (x_vals - x0) + y0

plt.figure(figsize=(6,4))
plt.plot(x_vals, y_vals, label="f(x) = x^2")
plt.plot(x_vals, tangent, '--', label=f"Tangent at x={x0}")
plt.scatter([x0], [y0], color='red')
plt.legend()
plt.title("Derivative as Slope of Tangent Line")
plt.grid(True)
plt.show()
```

```python
# Cell 9: Partial Derivatives (Markdown)
"""
## 4) Partial Derivatives
For multivariable functions like f(x,y) = x² + y², we compute ∂f/∂x and ∂f/∂y.
"""
```

```python
# Cell 10: Partial Derivatives
x, y = sp.symbols('x y')
f_xy = x**2 + y**2

df_dx = sp.diff(f_xy, x)
df_dy = sp.diff(f_xy, y)

f_xy, df_dx, df_dy
```

```python
# Cell 11: Gradient Vector (Markdown)
"""
## 5) Gradient Vector
The gradient is a vector of partial derivatives: ∇f(x,y) = [∂f/∂x, ∂f/∂y].
"""
```

```python
# Cell 12: Gradient Vector
grad_f = [df_dx, df_dy]
grad_f
```

```python
# Cell 13: Gradient Descent (Markdown)
"""
## 6) Gradient in Optimization
We use gradient descent intuition: update x, y by subtracting learning_rate * gradient.
"""
```

```python
# Cell 14: Gradient Descent Example
# Gradient descent example on f(x,y) = x^2 + y^2
f = lambda x, y: x**2 + y**2
grad = lambda x, y: np.array([2*x, 2*y])

# Start point
point = np.array([3.0, 4.0])
learning_rate = 0.1

history = [point]
for i in range(20):
    g = grad(*point)
    point = point - learning_rate * g
    history.append(point)

history = np.array(history)
plt.figure(figsize=(6,6))
plt.plot(history[:,0], history[:,1], 'o-', label="Gradient Descent Path")
plt.scatter([0], [0], color='red', label="Minimum")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Gradient Descent on f(x,y)=x²+y²")
plt.grid(True)
plt.show()
```
